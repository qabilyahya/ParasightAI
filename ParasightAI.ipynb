```python

dataset_path = Path("/content/dataset/malaria_hematie_dataset")

rows = []
for label in ["parasitized", "uninfected"]:
    for p in (dataset_path / label).glob("*"):
        if p.suffix.lower() in [".png", ".jpg", ".jpeg"]:
            rows.append({"filepath": str(p), "label": label})

df = pd.DataFrame(rows)
print(df.head())
print(df["label"].value_counts())
```

```python
IMG_SIZE = (224, 224)

def load_image_pil(path):
    img = Image.open(path).convert("RGB")
    img = img.resize(IMG_SIZE)
    img = np.array(img, dtype=np.float32) / 255.0
    return img
```

```python
p0 = df["filepath"].iloc[0]
x0 = load_image_pil(p0)
print("1 image shape:", x0.shape, "| min/max:", x0.min(), x0.max())



```
